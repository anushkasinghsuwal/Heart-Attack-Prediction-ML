## Label Encoding and Mapping 

```python
df['age group'].unique()
```

```python
# Mapping the age group

# Create a mapping dictionary

age_mapping = {
    "[0-10)": 1,
    "[10-20)": 2,
    "[20-30)": 3,
    "[30-40)": 4,
    "[40-50)": 5,
    "[50-60)": 6,
    "[60-70)": 7,
    "[70-80)": 8,
    "[80-90)": 9,
    "[90-100)": 10
}

# Apply the mapping to your DataFrame
df['age group'] = df['age group'].map(age_mapping)
```

```python
#df['weight'].unique()
```

```python
# Checking the values in 'weight' and their frequency to decide what to do with '?', and for mapping other values

# df['weight'].mode().iloc[0]
#df['weight'].value_counts()
```

```python
# encoding gender

label_encoder = LabelEncoder()
df['gender'] = label_encoder.fit_transform(df['gender'])
```

```python
# replacing normal pressure in label with 0 and other conditions with 1

df['label'] = df['label'].replace('>5', 1)
df['label'] = df['label'].replace('<30', 1)
df['label'] = df['label'].replace('NO', 0)
```

```python
# mapping max_glu_serum and A1Cresult

a={np.nan:0,'Norm':100,'>200':200,'>300':300}
df["max_glu_serum"]=df["max_glu_serum"].map(a)

b={np.nan:0,'Norm':5,'>7':7,'>8':8}
df["A1Cresult"]=df["A1Cresult"].map(b)
```

```python
# encoding remaining columns

# Create a LabelEncoder instance
le = LabelEncoder()

# Encode each specified column
for col in ['cast', 'Med']:
    df[col] = le.fit_transform(df[col])

```

```python
# mapping test columns

tests_cols = ['metformin','repaglinide','nateglinide','chlorpropamide','glimepiride','acetohexamide','glipizide','glyburide',
 'tolbutamide','pioglitazone','rosiglitazone','acarbose','miglitol','troglitazone','tolazamide','insulin','glyburide-metformin',
'glipizide-metformin','glimepiride-pioglitazone','metformin-rosiglitazone','metformin-pioglitazone']
replace_words={'Up':10,'Down':-10,'Steady':0,'No':-20}
for col in tests_cols:
    df[col] = df[col].replace(replace_words)
```

```python
# filling all nan values with -1 because those were not encoded

#df = df.fillna(-1)

# filling na values with -1 would indicate the values are unknown. we didn't use
# 0 because the encoder had assigned 0 for other values
```

```python
# dropping the weight column
df = df.drop(columns=['weight'])
```

## Feature Engineering

```python
selected_features = ['age group', 'gender', 'cast', 'insulin', 'A1Cresult', 'max_glu_serum', 'Med']
```

```python
# Sampling and Randomizing the huge data before train-test splitting
df2 = df.sample(frac=0.25, random_state=42)


df2
#[selected_features]
```

## Train and Test Splitting

```python
X = df2[selected_features]
#.drop(columns='label')
#.drop(columns= ['label', 'diag_1', 'diag_2', 'diag_3'])
y = df2['label']
```

```python
df2['label'].value_counts()
```

```python
# Splitting the datasets into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50)
X_train.shape, X_test.shape
```

```python
# Standardizing

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

## Implementing ML models 

```python
# initialize models
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

logistic_model = LogisticRegression()
knn_model = KNeighborsClassifier(n_neighbors = 5)
svm_model = SVC(kernel = "linear", class_weight='balanced')
rf_model = RandomForestClassifier(n_estimators = 100, random_state=42, class_weight='balanced')
```

```python
models = {
    'SVM': SVC(),
    'Random Forest': RandomForestClassifier(),
    'KNN': KNeighborsClassifier(),
    'Logistic Regression': LogisticRegression()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {accuracy:.2f}")
```

## Model Evaluation

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming the SVM model has already been trained and predictions have been made
svm_model = models['SVM']
y_pred_svm = svm_model.predict(X_test)

# Calculate precision, recall, and F1-score for the SVM model
precision_svm = precision_score(y_test, y_pred_svm, average='weighted', zero_division=0)
recall_svm = recall_score(y_test, y_pred_svm, average='weighted', zero_division=0)
f1_svm = f1_score(y_test, y_pred_svm, average='weighted', zero_division=0)

print(f'SVM Precision: {precision_svm:.2f}')
print(f'SVM Recall: {recall_svm:.2f}')
print(f'SVM F1-Score: {f1_svm:.2f}')
```
